```python
from sqlalchemy import create_engine

# Define SQL Server connection details
server = "localhost\SQLEXPRESS"  # The name of your server
database = 'master'  # The database you want to connect to

# Create a connection string with a trusted connection
connection_url = f"mssql+pyodbc://@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes"

# Create an engine
engine = create_engine(connection_url)

# Transfer data to SQL Server
table_name = 'telecom_dataset'
df.to_sql(table_name, engine, if_exists='replace', index=False)

# Confirm completion
print(f"Dataset transferred to SQL Server table: {table_name}")

# Dispose the engine
engine.dispose()
```

```python
from sqlalchemy import create_engine, text

# Define connection details
server = r'localhost\SQLEXPRESS' # Use a raw string (r'') to handle the backslash properly
database = 'master'
connection_url = f"mssql+pyodbc://@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server&trusted_connection=yes"

# Create the engine
engine = create_engine(connection_url)

# Test the connection by querying the table
with engine.connect() as conn:
    query = text("SELECT TOP 10 * FROM telecom_dataset") 
    
    result = conn.execute(query)
    for row in result:
        print(row)

engine.dispose()
```
